# Filtering the Data with Pandas

Filtering is one of the main things to look at when working with pandas because it is basically how we start every project by filtering the data that we want from the data and that we don'want.

```python
data
```

Now we'll filter out the rows with the highest number of sales.


```python
highest_sales = (data['Sales'] > 300)
```

```python
highest_sales #dor each values of sales which is greater than 300 the it is going to be True and if the value is less than 300 it is going to be False
```

```python
data.loc[highest_sales] #now we can see that all the values of sales from our data frame are over or greater than 300
```

```python
data.loc[highest_sales, ['Product', 'Profit', 'Sales']] 
#when i run this cellwe will get only the values of sales grester than 300 with Product and Profit column  
```

above we simply filtered our sales data based on the categories of products and profit we have

```python
data['Product'].unique() #after running this cell we will get to know how many categores we have for product
```
